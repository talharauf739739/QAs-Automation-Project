**Verifying LLMs Answer's with Orignal Answer, And Update Answer**

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# Assuming 'generated_text' is defined somewhere in your code
generated_text = generated_text

if "Sorry, I don't have an answer for that question." in generated_text:
    # If the generated_text contains the placeholder string, assign nothing to row_with_max_similarity
    row_with_max_similarity = None
else:
    # Read the Excel file
    detect = pd.read_excel('/content/Q_A responses_against_llms_Keywords.xlsx')

    # Combine generated_text and Find_Matched_Answer for TF-IDF computation
    text_corpus = [generated_text] + list(detect['Find_Matched_Answer'])

    # Compute TF-IDF vectors
    tfidf_vectorizer = TfidfVectorizer()
    tfidf_matrix = tfidf_vectorizer.fit_transform(text_corpus)

    # Calculate cosine similarity manually
    generated_text_vector = tfidf_matrix[0]
    jaccard_similarities = []
    for index, row in detect.iterrows():
        answer_vector = tfidf_matrix[index + 1]
        intersection = np.sum(generated_text_vector.multiply(answer_vector))
        union = np.sum(generated_text_vector) + np.sum(answer_vector) - intersection
        jaccard_sim = intersection / union
        jaccard_similarities.append(jaccard_sim)

    # Find the row with the highest Jaccard similarity
    max_similarity_index = np.argmax(jaccard_similarities)
    row_with_max_similarity = detect.iloc[max_similarity_index]

print(row_with_max_similarity)
```

```python
#Duplicate Case
#Database Creation And Checking For Previous Questions

Duplicate_check = ''

# Initialize an empty list to store recent questions
question_database = []

# Function to check for duplicate questions
def is_duplicate(question):
    return question in question_database

# Loop through QA_input to process each item
for item in QA_input:
    question = item['question']

    # Check if the question is a duplicate
    if is_duplicate(question):
        Duplicate_check = 'Onclick Function on Duplicate'
    else:
        # Add the question to the recent questions database
        question_database.append(question)

        # If the number of recent questions exceeds 50, remove the oldest question
        if len(question_database) > 50:
            question_database.pop(0)

        Duplicate_check = 'Nothing to do here'

    print(Duplicate_check)

```
